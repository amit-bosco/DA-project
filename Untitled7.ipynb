```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('/content/amazon.csv')

# Remove currency symbols and commas
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','').astype(float)

# Convert rating_count
df['rating_count'] = df['rating_count'].str.replace(',','').astype(float)

# Convert discount %
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)

# Handle missing ratings
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop duplicates
df = df.drop_duplicates()

# Save cleaned dataset
df.to_csv('/content/amazon_cleaned.csv', index=False)

df.head()
```

```python
import sqlite3

conn = sqlite3.connect('amazon.db')
cursor = conn.cursor()

cursor.execute('''
SELECT product_name, rating_count
FROM amazon
ORDER BY rating_count DESC
LIMIT 10;
''')

results = cursor.fetchall()
for row in results:
    print(row)

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('amazon.db')
cursor = conn.cursor()

cursor.execute('''
SELECT main_category, AVG(rating) as avg_rating
FROM amazon
GROUP BY main_category
ORDER BY avg_rating DESC;
''')

results = cursor.fetchall()

for row in results:
    print(row)

conn.close()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert the results into a pandas DataFrame for easier plotting
avg_ratings_df = pd.DataFrame(results, columns=['main_category', 'avg_rating'])

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(avg_ratings_df['main_category'], avg_ratings_df['avg_rating'], color='skyblue')
plt.xlabel('Main Category')
plt.ylabel('Average Rating')
plt.title('Average Rating by Main Category')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.ylim(3.8, 4.4) # Set a reasonable y-axis limit to highlight differences
plt.tight_layout()
plt.show()
```

```python
import sqlite3

conn = sqlite3.connect('amazon.db')
cursor = conn.cursor()

cursor.execute('''
SELECT product_name, discounted_price, rating
FROM amazon
WHERE discounted_price < 500 AND rating > 4;
''')

results = cursor.fetchall()

for row in results:
    print(row)

conn.close()
```

```python
filtered_products_df = pd.DataFrame(results, columns=['product_name', 'discounted_price', 'rating'])

print("Filtered Products DataFrame created successfully. Displaying the first 5 rows:")
filtered_products_df.head()
```

```python
merged_df = pd.merge(filtered_products_df, df[['product_name', 'main_category']], on='product_name', how='left')

category_counts = merged_df['main_category'].value_counts()

print("Main category counts for filtered products:")
print(category_counts)
```

```python
print("Descriptive statistics for discounted price and rating of filtered products:")
filtered_products_df[['discounted_price', 'rating']].describe()
```

```python
import matplotlib.pyplot as plt

# 1. Histogram for 'discounted_price'
plt.figure(figsize=(10, 6))
plt.hist(filtered_products_df['discounted_price'], bins=20, edgecolor='black')
plt.title('Distribution of Discounted Price (< 500)')
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Number of Products')
plt.grid(axis='y', alpha=0.75)
plt.show()

# 2. Histogram for 'rating'
plt.figure(figsize=(10, 6))
plt.hist(filtered_products_df['rating'], bins=10, edgecolor='black')
plt.title('Distribution of Ratings (> 4)')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.grid(axis='y', alpha=0.75)
plt.show()

# 3. Bar chart for 'main_category' counts
plt.figure(figsize=(12, 7))
category_counts.plot(kind='bar', color='lightcoral')
plt.title('Main Category Counts for Filtered Products')
plt.xlabel('Main Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
